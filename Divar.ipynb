```python
url = soup.html.body.find(id="app").main.find('div', attrs={'class': 'post-list-grid-_-eef81 virtual-infinite-scroll'}).find('div', attrs={'class': 'virtual-infinite-scroll__viewport'}).find('div', attrs={'class': 'browse-post-list-_-f3858'}).find_all('div', attrs={'class': 'post-card-item-_-af972 kt-col-6-_-bee95 kt-col-xxl-4-_-e9d46'})

for i in url:
    try:
        link="https://divar.ir"+i.a['href']
        site = requests.get(link)
        TopicExtract = BeautifulSoup(site.text, 'html.parser')

        Topic = TopicExtract.html.body.find(id="app").find('div', attrs={'class': 'container-_-b6963 kt-container'}).find('div', attrs={'class': 'kt-page-title__title kt-page-title__title--responsive-sized'}).text
        Price = TopicExtract.html.body.find(id="app").find('div', attrs={'class': 'container-_-b6963 kt-container'}).find('p', attrs={'class': 'kt-unexpandable-row__value'}).text

        print(Topic,Price)
    except:
        pass
```

### Print Advertising inthe first Page when the price="توافقی"

```python
url = soup.html.body.find(id="app").main.find('div', attrs={'class': 'post-list-grid-_-eef81 virtual-infinite-scroll'}).find('div', attrs={'class': 'virtual-infinite-scroll__viewport'}).find('div', attrs={'class': 'browse-post-list-_-f3858'}).find_all('div', attrs={'class': 'post-card-item-_-af972 kt-col-6-_-bee95 kt-col-xxl-4-_-e9d46'})

for i in url:
    try:
        a_dict = {}
        link="https://divar.ir"+i.a['href']
        site = requests.get(link)
        TopicExtract = BeautifulSoup(site.text, 'html.parser')

        Topic = TopicExtract.html.body.find(id="app").find('div', attrs={'class': 'container-_-b6963 kt-container'}).find('div', attrs={'class': 'kt-page-title__title kt-page-title__title--responsive-sized'}).text
        Price = TopicExtract.html.body.find(id="app").find('div', attrs={'class': 'container-_-b6963 kt-container'}).find('p', attrs={'class': 'kt-unexpandable-row__value'}).text
       
        a_dict[Topic] = Price
        
        for key , value in a_dict.items():
            if value=="توافقی":
                print(a_dict)
        
        
    except:
        pass
```
